````python
#Create a function to get the percentage of null values of any column
def get_percentage_of_null_values(df, column_name):
    """Calculates the percentage of null values in a specified column of a DataFrame.

    Args:
        df: pandas DataFrame.
        column_name: Name of the column.

    Returns:
        float: The percentage of null values in the column.
    """

    # Calculate the number of null values in the column
    num_nulls = df[column_name].isnull().sum()

    # Calculate the total number of rows in the DataFrame
    total_rows = len(df)

    # Calculate the percentage of null values
    percentage_nulls = (num_nulls / total_rows) * 100

    return percentage_nulls

# Example usage:
column_to_check = 'seasons'  # Replace with the actual column name
percentage = get_percentage_of_null_values(movies_tv_shows_df, column_to_check)
print(f"Percentage of null values in '{column_to_check}': {percentage:.2f}%")
````

Age Certification null value handling

````python
# get the count of null values of age certification
movies_tv_shows_df['age_certification'].isnull().sum()
````

````python
# Get the most frequent value from the DataFrame
most_frequent_value = movies_tv_shows_df['age_certification'].value_counts().idxmax()

print(most_frequent_value)
````

````python
# check for the percentage of null values
percentage = get_percentage_of_null_values(movies_tv_shows_df, 'age_certification')

#if percentage of null values are more than 5 then replace the null values otherwise drop them
if percentage > 5:
  movies_tv_shows_df['age_certification'].fillna(most_frequent_value, inplace=True)
else:
  movies_tv_shows_df.dropna(subset = ['age_certification'], inplace=True)
````

````python
# check all the unique values of age restriction column
movies_tv_shows_df['age_certification'].unique()
````

````python
# convert the column to numeric column
movies_tv_shows_df['seasons'] = pd.to_numeric(movies_tv_shows_df['seasons'], errors='coerce')
````
